### Abstract
The goal of this project is to use a machine learning model to predict the presence of  the icelanding rock ptarmagin in a specific region of iceland. To do so we will be using data from ... will be used in conjungtion with several machine learning models to generate a relative index of occourance map. This map will tell us where we can expect the  Ptarmagin habitat to be, which will in turn allow us better insights on how might preserve the habitat and animal species.

### Methods
For this model we will be generating an ensemble model, not only do they generally perform better than single models(Delgado Paper) but it will also provide ample opportunity to showcase what we have learned throughout the semester. Our model will be constructed by averaging the RIO maps of Random Forest, Gradient Boosting, and MaxEnt models. This model will also provide an interesting benchmark into how these different models perform with ecological data. We have seen from papers in class that both Gradient Boosting and MaxEnt have been used for this type of application. \\

### Generating Closest Test Prediction
We want to be able to evaluate the performance of our model with the testing data. 
To do so we need to generate the input data for each of our entry data, so for each 
point in our testing data we will find the closest corresponding point in our prediction lattice. 
Once we genratete our predictions from the model we will compare them to our testing data. 

```python
PresenceInputDataIndex = []
for i in range(len(Presence)):
    xPres = Presence['x'][i]
    yPres = Presence['y'][i]
    xPredict = Lattice['POINT_X']
    yPredict = Lattice['POINT_Y']
    distance = (((xPredict - xPres)**2) + ((yPredict - yPres)**2)) ** .5
    PresenceInputDataIndex.append(distance[distance == min(distance)].index[0])
```

```python
TestData = Lattice.drop(['POINT_X','POINT_Y', 'Latitude', 'Longitude'], axis=1).loc[PresenceInputDataIndex]
TestData['Latitude'] = Presence['Latitude'].tolist()
TestData['Longitude'] = Presence['Longitude'].tolist()
TestData['PresBack'] = 1
TestData.head()
```
